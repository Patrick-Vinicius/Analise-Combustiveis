2) Realizar a conversão do tipo string para o tipo datetime das colunas de DATA INICIAL E DATANA FINAL

11) Série Temporal com a diferença com a diferença absoluta e percentual entre os valores mais baratos e caros. Apresente também ao lado os estados com maiores e menores preços

```python
df_max =df2.groupby("ANO-MES").max()["PREÇO MÉDIO REVENDA"]
df_min =df2.groupby("ANO-MES").min()["PREÇO MÉDIO REVENDA"]

idx_max =df2.groupby("ANO-MES")["PREÇO MÉDIO REVENDA"].idxmax()
idx_min =df2.groupby("ANO-MES")["PREÇO MÉDIO REVENDA"].idxmin()
```

```python
df2.groupby("ANO-MES")["PREÇO MÉDIO REVENDA"].idxmax()
```

```python
df_diff = pd.DataFrame()
```

```python
df_diff["Diferença Absoluta"] = df_max - df_min
df_diff["Diferença de %"] = (df_max - df_min) / df_min * 100
df_diff["Max"] = df_max
df_diff["Min"] = df_min
```
